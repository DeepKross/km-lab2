```python
import numpy as np

# Визначаємо функцію f(x)
def f(x):
    return 3*x**4 - 2*x**2 + x + 1

# Тестуємо функцію на декількох значеннях
test_values = np.array([0, 0.5, 1])
f_values = f(test_values)
f_values
```

```python
from numpy.polynomial import Polynomial

# Використовуємо метод найменших квадратів для побудови поліноміального наближення
x = np.linspace(0, 1, 100)
y = f(x)

# Знаходимо коефіцієнти многочлена
coefficients = np.polyfit(x, y, 3)
polynomial = Polynomial(coefficients)

# Обчислюємо значення многочлена для тестових значень
poly_values = polynomial(test_values)
coefficients, poly_values
```

```python
# Обчислюємо відхилення між функцією та многочленом
deviation = np.abs(f(x) - polynomial(x))

# Знаходимо максимальне відхилення на інтервалі
max_deviation = np.max(deviation)

max_deviation
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='Original function: $f(x)$', color='blue')
plt.plot(x, polynomial(x), label='Approximation: $P(x)$', color='red', linestyle='--')
plt.fill_between(x, f(x), polynomial(x), color='grey', alpha=0.3, label='Deviation')
plt.title("Original Function vs Polynomial Approximation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
